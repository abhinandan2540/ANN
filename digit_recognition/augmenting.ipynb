```python

import os
import Augmentor

root_dir='digits'

def digit_augmentor(root_dir):
    files=os.listdir(root_dir)
    for file in files:
        file_path=os.path.join(root_dir, file)

        P=Augmentor.Pipeline(file_path)
        # P.zoom_random(probability=0.5,percentage_area=0.5)
        P.rotate(probability=0.5, max_left_rotation=15, max_right_rotation=15)
        P.shear(probability=0.5, max_shear_left=15, max_shear_right=15)
        P.sample(100)

digit_augmentor(root_dir=root_dir)
```

Building a Multi Layer Perceptron with 2 `hidden layer`

`1 input layer (1 node)`, `1st hidden layer (10 neuron)`, `2nd hidden layer (10 neuron)`, `1 output layer (1 node)` 

declaring `Input` and `Output` values

```python
import numpy as np
np.random.seed(42)

X=np.random.rand(1).reshape(1,1)  # X of input node
Y_true=1 # actual Y value of output node

print(X)
print(X.shape)
print(Y_true)
```

weights and bias for `Input` to `Hidden layer 1`

```python
np.random.seed(42)
num_inputs=1
num_nodes=10

# weights for Input Node -> Hidden Layer 1
# W_input_hidden1=np.random.rand(1,10) 
W_input_hidden1=np.random.uniform(low=-1.0, high=1.0,size=(num_inputs, num_nodes))

# b_input_hidden1=np.random.rand(1,10)
b_input_hidden1=np.random.uniform(low=-1.0, high=1.0, size=(1, num_nodes))

print(W_input_hidden1)
print(W_input_hidden1.shape)
print(b_input_hidden1)
print(b_input_hidden1.shape)
```

weights and bias for `Hidden Layer 1` to `Hidden Layer 2`

```python
np.random.seed(42)

# Weights for hidden layer 1 to hidden layer 2
W_hidden1_hidden2=np.random.uniform(low=-1.0, high=1.0, size=(num_nodes, num_nodes))
b_hidden1_hidden2=np.random.uniform(low=-1.0, high=1.0, size=(1, num_nodes))

print(W_hidden1_hidden2)
print(W_hidden1_hidden2.shape)
print(b_hidden1_hidden2)
print(b_hidden1_hidden2.shape)

```

Weights and bias for `Hidden Layer 2` to `Output Layer`

```python
np.random.seed(42)

# weights for Hidden layer 2 to Output Layer
W_hidden2_output=np.random.uniform(low=-1.0, high=1.0, size=(num_nodes,1))

# bias for Hidden Layer 2 to Output Layer
b_hidden2_output=np.random.rand(1).reshape(1,1)
# b_hidden2_output=np.random.uniform(low=-1.0, high=1.0, size=(num_nodes,1))

print(W_hidden2_output)
print(W_hidden2_output.shape)
print(b_hidden2_output)
print(b_hidden2_output.shape)
```

Activation function : `Sigmoid`

```python
def Sigmoid_Function(z):
    return 1/(1+np.exp(-z))
```

*under the hood calculations*

*from Input Layer to Hidden Layer 1*

```python
# from input layer to hidden layer 1
Z_input_hidden1=X@W_input_hidden1+b_input_hidden1
print(Z_input_hidden1)

A_hidden1=Sigmoid_Function(Z_input_hidden1)
print(A_hidden1)
print(A_hidden1.shape)


```

*from Hidden Layer 1 to Hidden Layer 2*

```python
# from hidden layer 1 to hidden layer 2
Z_hidden1_hidden2 = A_hidden1@W_hidden1_hidden2+b_hidden1_hidden2
print(Z_hidden1_hidden2)
print(Z_hidden1_hidden2.shape)

A_hidden2 = Sigmoid_Function(Z_hidden1_hidden2)
print(A_hidden2)
print(A_hidden2.shape)
```

*from Hidden Layer 2 to Output Layer*

```python
# from hidden layer 2 to output layer
Z_hidden2_output = A_hidden2@W_hidden2_output+b_hidden2_output
print(Z_hidden2_output)
print(Z_hidden2_output.shape)

Y_pred = Sigmoid_Function(Z_hidden2_output)
print(Y_pred)
print(Y_pred.shape)
```

```python
# shapes information
# input layer
print(X.shape)

# input layer -> hidden layer 1
print(W_input_hidden1.shape)
print(b_input_hidden1.shape)

# hidden layer 1 -> hidden layer 2
print(W_hidden1_hidden2.shape)
print(b_hidden1_hidden2.shape)

# hidden layer 2 -> output layer
print(W_hidden2_output.shape)
print(b_hidden2_output.shape)

```

Training Loop

```python
epochs=1000000
Learning_Rate=0.01 # learning rate

for epoch in range(epochs):

    # FORWARD PASS
    # Input -> Hidden Layer 1
    # X (1,1)
    # W_input_hidden1 (1,10)
    # b_input_hidden1 (1,10)
    Z_input_hidden1 = X@W_input_hidden1+b_input_hidden1
    A_hidden1 = Sigmoid_Function(Z_input_hidden1) # A_hidden1 (1,10)


    # # Hidden Layer 1 -> Hidden Layer 2
    # A_hidden1 (1,10)
    # W_hidden1_hidden2 (10,10)
    # b_hidden1_hidden2 (1,10)
    Z_hidden1_hidden2 = A_hidden1@W_hidden1_hidden2+b_hidden1_hidden2
    A_hidden2 = Sigmoid_Function(Z_hidden1_hidden2) # A_hidden2 (1,10)


    # # Hidden Layer 2 -> Output Layer
    # A_hidden2 (1,10)
    # W_hidden2_output (10,1)
    # b_hidden2_output (1,1)
    Z_hidden2_output = A_hidden2@W_hidden2_output+b_hidden2_output
    Y_pred = Sigmoid_Function(Z_hidden2_output) # Y_pred (1,1)

    # LOSS FUNCTION
    L=0.5*(Y_pred - Y_true)**2

    # BACK PROPAGATION
    # OUTPUT LAYER -> HIDDEN LAYER 2

    # Weights Gradients
    # dL/dW_hidden2_output = (dL/dY_pred) * (dY_pred/dZ_hidden2_output) * (dZ_hidden2_output/dW_output_hidden2)
    dL_dY_pred=(Y_pred - Y_true) # (1,1)
    dY_pred_dZ_hidden2_output=Y_pred * (1-Y_pred) # (1,1)
    dZ_dW_hidden2_output=A_hidden2.T # (10,1)

    # dL/dZ_hidden2_output = (dL/dY_pred) * (dY_pred/dZ_hidden2_output)
    dL_dZ_hidden2_output= (dL_dY_pred) @ (dY_pred_dZ_hidden2_output) # (1,1)

    # dL/dW_output_hidden2 = (dL/dZ_hidden2_output) * (dZ_hidden2_output/dW_hidden2_output)
    dL_dW_hidden2_output=(dZ_dW_hidden2_output) @ (dL_dZ_hidden2_output) # (10,1)

    # Bias Gradients
    # dL/db_hidden2_output = (dL/dY_pred) * (dY_pred/dZ_hidden2_output) * (dZ_hidden2_output/db_hidden2_output)
    dZ_db_hidden2_output=1
    dL_db_hidden2_output=(dL_dZ_hidden2_output) * (dZ_db_hidden2_output) # (1,1)



    # HIDDEN LAYER 2 -> HIDDEN LAYER 1
    # Weights Gradients
    # dL/dW_hidden1_hidden2 = (dL/dA_hidden2) * (dA_hidden2/dZ_hidden1_hidden2) * (dZ_hidden1_hidden2/dW_hidden1_hidden2)
    # dL/dA_hidden2 = (dL/dY_pred) * (dY_pred/dZ_hidden2_output) * (dZ_hidden2_output/dA_hidden2)
    # dA_hidden2/dZ_hidden1_hidden2 = A_hidden2 * (1 - A_hidden2)
    # dZ_hidden1_hidden2/dW_hidden1_hidden2 = A_hidden1

    # dL/dY_pred = (Y_pred - Y_true)
    # dY_pred/dZ_hidden2_output = Y_pred * (1 - Y_pred)
    # dZ_hidden2_output/dA_hidden2 = W_hidden2_output

    # dL/dA_hidden2 = (dL/dY_pred) * (dY_pred/dZ_hidden2_output) * (dZ_hidden2_output/dA_hidden2)
    dZ_hidden2_output_dA_hidden2 = W_hidden2_output # (10,1)
    # dL_dA_hidden2 = (dZ_hidden2_output_dA_hidden2) @ (dL_dZ_hidden2_output) # (10,1)
    dL_dA_hidden2 = (dL_dZ_hidden2_output) @ (dZ_hidden2_output_dA_hidden2.T) # (1,10)


    # dA_hidden2/dZ_hidden1_hidden2 = A_hidden2 * (1 - A_hidden2)
    dA_hidden2_dZ_hidden1_hidden2 = A_hidden2 * (1-A_hidden2)  # (1,10)

    # dL_dZ_hidden1_hidden2 = (dL/dA_hidden2)*(dA_hidden2/dZ_hidden1_hidden2)
    dL_dZ_hidden1_hidden2 = (dL_dA_hidden2) * (dA_hidden2_dZ_hidden1_hidden2) # (1,10)

    # dZ_hidden1_hidden2/dW_hidden1_hidden2 = A_hidden1
    dZ_hidden1_hidden2_dW_hidden1_hidden2 = A_hidden1 # (1,10)

    # dL/dW_hidden1_hidden2 = (dL/dA_hidden2) * (dA_hidden2/dZ_hidden1_hidden2) * (dZ_hidden1_hidden2/dW_hidden1_hidden2)
    dL_dW_hidden1_Hidden2 = (dZ_hidden1_hidden2_dW_hidden1_hidden2.T) @ (dL_dZ_hidden1_hidden2) # (10,10)


    # Bias Gradients
    # dL/db_hidden1_hidden2 = (dL/dA_hidden2) * (dA_hidden2/dZ_hidden1_hidden2) * (dZ_hidden1_hidden2/db_hidden1_hidden2)
    dZ_hidden1_hidden2_db_hidden1_hidden2 = 1
    dL_db_hidden1_hidden2 = (dL_dZ_hidden1_hidden2) * (dZ_hidden1_hidden2_db_hidden1_hidden2) # (1,10)



    # HIDDEN LAYER 1 -> INPUT LAYER
    # Weights Gradients
    # dL/dW_input_hidden1 = (dL/dA_hidden1) * (dA_hidden1/dZ_input_hidden1) * (dZ_input_hidden1/dW_input_hidden1)
    # dL/dA_hidden1 = (dL/dA_hidden2) * (dA_hidden2/dZ_hidden1_hidden2) * (dZ_hidden1_hidden2/dA_hidden1)
    # dA_hidden1/dZ_input_hidden1 = A_hidden1 * (1 - A_hidden1)
    # dZ_input_hidden1/dW_input_hidden1 = X

    # dL/dA_hidden1 = (dL/dA_hidden2) * (dA_hidden2/dZ_hidden1_hidden2) * (dZ_hidden1_hidden2/dA_hidden1)
    # dL/dA_hidden2 (1,10)
    # dA_hidden2/dZ_hidden1_hidden2 (1,10)
    # dZ_hidden1_hidden2/dA_hidden1 = W_hidden1_hidden2 (10,10)
    dZ_hidden1_hidden2_dA_hidden1 = W_hidden1_hidden2 # (10,10)

    # dL/dZ_hidden1_hidden2 = (dL/dA_hidden2) * (dA_hidden2/dZ_hidden1_hidden2) (1,10)
    # dL/dA_hidden1
    dL_dA_hidden1 = dL_dZ_hidden1_hidden2 @ dZ_hidden1_hidden2_dA_hidden1.T # (1,10)

    dA_hidden1_dZ_input_hidden1 = A_hidden1 * (1-A_hidden1) # (1,10)

    # dL/dZ_input_hidden1 = (dL/dA_hidden1) * (dA_hidden1/dZ_input_hidden1)
    dL_dZ_input_hidden1 = (dL_dA_hidden1) * (dA_hidden1_dZ_input_hidden1) # (1,10)

    dZ_input_hidden1_dW_input_hidden1 = X # (1,1)

    # dL/dW_input_hidden1  = (dL/dZ_input_hidden1) * (dZ_input_hidden1/dW_input_hidden1)
    dL_dW_input_hidden1 = dZ_input_hidden1_dW_input_hidden1.T @ dL_dZ_input_hidden1 # (1,10)


    # Bias Gradients
    # dL/db_input_hidden1 = (dL/dA_hidden1) * (dA_hidden1/dZ_input_hidden1) * (dZ_input_hidden1/db_input_hidden1)
    # dL/dZ_input_hidden1 = (dL/dA_hidden1) * (dA_hidden1/dZ_input_hidden1)
    # (dZ_input_hidden1/db_input_hidden1)
    dZ_input_hidden1_db_input_hidden1 = 1
    dL_db_input_hidden1 = dL_dZ_input_hidden1 * dZ_input_hidden1_db_input_hidden1 # (1,10)



    # WEIGHTS AND BIAS UPGRADATIONA
    # OUTPUT LAYER -> HIDDEN LAYER 2
    W_hidden2_output -= Learning_Rate*(dL_dW_hidden2_output)  # (10,1)
    b_hidden2_output -= Learning_Rate*(dL_db_hidden2_output)  # (1,1)

    # HIDDEN LAYER 2 -> HIDDEN LAYER 1
    W_hidden1_hidden2 -= Learning_Rate*(dL_dW_hidden1_Hidden2)  # (1,10)
    b_hidden1_hidden2 -= Learning_Rate*(dL_db_hidden1_hidden2)  # (1,10)

    # HIDDEN LAYER 1 -> INPUT LAYER
    W_input_hidden1 -= Learning_Rate*dL_dW_input_hidden1
    b_input_hidden1 -= Learning_Rate*dL_db_input_hidden1

    if (epoch+1)%100==0:
        print(f"Epoch : {epoch+1}| Loss : {L.item():.6f} | Y_pred : {Y_pred.item():.6f}")

```
